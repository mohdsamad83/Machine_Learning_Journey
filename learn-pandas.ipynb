# Aggregate statistics

```python
# Group by 'Sex' and calculate the mean for numeric columns only
df = pd.read_csv('train.csv')
df.groupby('Sex').mean(numeric_only=True)
means = df.groupby('Sex').mean(numeric_only=True)

totals = means.sum().to_frame().T
totals.index = ['Total']
result = pd.concat([means, totals])
print(result)
```
